### Kmeans clustering

Kmeans clustering is part of the unsupervised branch of machine learning. This means that we don't have a defined target variable to work or aim towards like we would in traditional regression or classification tasks. This means that the main objective of this method therefore is to be able to identify distinct groups of objects that share similar characteristics, for example groups of shoppers, groups of films or groups of stores. 

The way this method works is by firstly defining a target number of groups to create, k, which the algorithm will use to create centroids of groups from the data. From these centroids, every data point is llocated to each of the clusters by reducing the in-cluster sum of squares (the distance between the centroid and the individual point). It will start off with a group of randomly seected centroids which are used for the beggining poins of each cluster. It then performs iterative calculations, changing the position of the centroid, to optimise the the position of the centroids for the given number of clusters. It is deemed as optimised either when the centroids do not change with each iteration, or the defined number of iterations has been completed.

The main issue with this is since this is an unsupervised machine learning algorithm and we don't always know what is the optimal number of clusters for the model to use when we start. To this end therefore there are different ways of identifying the optimal number of clusters including the "elbow method" and finding the maximum silhouette score by the number of clusters. Furthermore, since these two techniques don't always agree with each other perfectly, and clustering can be subjective, we need to evaluate the results in relation to the actual values underlying the clusters to see if they make sense. We will attempt to show you how this works below.

However, one key thing to note is that the scales of the two target variables are significantly different where teh residential population is of orders of magnitude large than the percentage of residents of class 1. Since kmeans clustering is a distance based algorithm, we need to ensure that the values are within roughly the same range and same scale. We can thus do this using the suite of scalers and normalising algorithms within sklearn, but for our purposes we can use the robust scaler as follows:

```python
#import the necessary scaler
from sklearn.preprocessing import RobustScaler

#create the scaler
scaler = RobustScaler()

#create data to transform on
tr_data = stores_dataset[["id", "retailer", "fascia",
                         "store_name", "residential_pop",
                         "class_1_pop"]].copy()

#set the columns to transofmr
cols_transform = ["residential_pop",
                 "class_1_pop"]

#fit the algorithm to the data
for col in cols_transform:
    tr_data[col] = scaler.fit_transform(tr_data[col].values.reshape(-1,1))
```

```python
#reapet the plot as before
fig = plt.figure(figsize = (8,8))
gs = GridSpec(4,4)

ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_x = fig.add_subplot(gs[0, 0:3])
ax_hist_y = fig.add_subplot(gs[1:4, 3])


ax_scatter.scatter(
    tr_data["residential_pop"],
    tr_data["class_1_pop"])

ax_hist_x.hist(tr_data["residential_pop"])
ax_hist_y.hist(tr_data["class_1_pop"],
              orientation = "horizontal")

plt.tight_layout()
```

We can now see that both of our target variables are primarily on the same scale and range which means that we can then start to run the kmeans clustering. 

But wait, didn't we say at the beginning that we don't know how many clusters to implement at the beginning? Correct! We can see from above that there are no necessary clear groups and thus we have to find out what the optimal value may be.

The two main methods for this are the "Elbow plot" method and the "Silhouette Score" method. The first of these methods uses the <b>inertia</b> in the clusters which is the sum of squared distances of samples to their closest cluster centre to find the inflection point where the inertia value begins to flatten out (like an elbow) wich suggests that the optimal number of clusters has been reached. You could also use the Distortion value in this case as well (the average squared distances from the cluster centres of the respective clusters). The second method seeks to find where the Silhouette score is at the greatest value. Thsi is because the score is a measure of how similar an object is to its own cluster compared to other clusters, with values closer to 1 suggesting better cluster delineation. Thus we look for the greatest value.

We can thus implement these as follows:

```python
#import the necessary libraries
from sklearn import metrics
from sklearn.cluster import KMeans

#create an empty list
list_SSE = []
#set ther ange of clusters to evaluate
min_k = 1
max_k = 10
range_k = range(min_k, max_k)

#iterate over the range
for i in range_k:
    #perform the clustering algorithm
    km = KMeans(n_clusters = i,
               init = "random",
               n_init = 10,
               max_iter = 300,
               tol = 1e-04, 
                random_state = 22)
    #fit this to the data
    km.fit(tr_data[["residential_pop",
                 "class_1_pop"]])
    #add the SEE score
    list_SSE.append(km.inertia_) 
```

```python
#create the base plots
fig, ax = plt.subplots(1,1, figsize = (10,10))

#add the data to the axis
ax.plot(range_k, list_SSE, marker = "o")

#add the necessary labels
ax.set_xlabel("Number of Clusters", fontsize = 20, labelpad = 20)
ax.set_ylabel("SSE", fontsize =20, labelpad = 20)
ax.set_title("Kmeans Elbow plot", fontsize = 25, pad = 20)

plt.show()
```

```python
#for the silhouette score
import random

#create an empty list
silhouette = []

#iteratre over the number of clusters to evaluate
for i in range(2,10):
    
    #create an empty list to hold the averages
    average = []
    #perform the clustering algorithm several times for each number of clusters
    for x in range(1,10):
        #set the number of clusters
        k_cluster = i
        #generate a arandom seed number
        random_seed = random.randint(1,101)
        #apply the KMeans clustering algorithm
        kmeans_method = KMeans(n_clusters = k_cluster,
                              random_state = random_seed)
        kmeans_method.fit(tr_data[["residential_pop",
                 "class_1_pop"]])
        #extract the labels
        labels = kmeans_method.labels_
        #extract the silhouette score
        a = metrics.silhouette_score(tr_data[["residential_pop",
                 "class_1_pop"]], labels)
        #append the result
        average.append(a)
    #clauclate the average silhouette score for each number of clusters 
    silhouette.append(sum(average)/len(average))
```

```python
#create a base axis
fig, ax = plt.subplots(1,1, figsize = (10,10))

#plot the results
ax.plot(range(2,10), silhouette, marker = "o")

#add the lbels
ax.set_xlabel("Number of Clusters", fontsize = 20, labelpad = 20)
ax.set_ylabel("Silhoute score", fontsize =20, labelpad = 20)
ax.set_title("Kmeans silhouette plot", fontsize = 25, pad = 20)

plt.show()
```

```python
#combine both plots
fig, ax = plt.subplots(1,2, figsize = (15,8))

ax[0].plot(range(2,10), silhouette, marker = "o")
ax[0].set_xlabel("Number of Clusters", fontsize = 20, labelpad = 20)
ax[0].set_ylabel("Silhoute score", fontsize =20, labelpad = 20)
ax[0].set_title("Kmeans silhouette plot", fontsize = 25, pad = 20)

ax[1].plot(range_k, list_SSE, marker = "o")
ax[1].set_xlabel("Number of Clusters", fontsize = 20, labelpad = 20)
ax[1].set_ylabel("SSE", fontsize =20, labelpad = 20)
ax[1].set_title("Kmeans Elbow plot", fontsize = 25, pad = 20)
```

What these two plots appear to suggest to use is that the optimal number of clusters to implement with this dataset is three clusters (highest silhouette score and beginning of the elbow in the elbow plot). Of course, although the model suggests these results, we need to actually verify whether they make sense to us.

We can thus implement this as follows:

```python
#create the model
kmeans_method = KMeans(n_clusters = 3,
                      random_state = 42,
                      n_init = 10)
#fit it to the data
kmeans_method.fit(tr_data[["residential_pop",
                 "class_1_pop"]])
#extract the labels
tr_data["labels"] = kmeans_method.labels_ + 1
```

We then need to see what these results actually look like. Firstly we can see what they look like on the transformed data:

```python
#create the base axis
fig, ax = plt.subplots(1,1, figsize = (8,8))

#create the scatter plot
sns.scatterplot(data = tr_data,
               x = "residential_pop",
               y = "class_1_pop",
               hue = "labels",
               palette = "tab10" )

ax.legend(fontsize = 20,
         title_fontsize = 20,
         title = "Labels")

ax.set_ylabel("Class 1 population",
             fontsize = 20,
             labelpad = 20)
ax.set_xlabel("Residential population",
             fontsize = 20,
             labelpad = 20)
```

We can thus see what the clusters look like on the transformed data and while they are not clear cut clusters in terms of the data itself, the model suggests that these are the optimal number of clusters.

But what about the centre, where did these clutsers originate from? We can extract this value using the `cluster_centeres_` attribute which we can then plot ontop of the original data:

```python
#extract the centres
centres = pd.DataFrame(kmeans_method.cluster_centers_,
                       columns = ["x", "y"])
```
